1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# gráfico para posse de imóvel
sns.pointplot(x='posse_de_imovel', y='renda', data=df)
plt.title('Renda média por posse de imóvel')
plt.ylabel('Renda média')
plt.xlabel('Possui imóvel')
plt.show()
````

-> Pessoas que possuem imóvel tendem a ganhar mais do que quem não possui.

-> Isso sugere que a variável 'posse_de_imovel' pode ser boa para prever renda.

````python
# Gráfico para posse de veículo
plt.figure(figsize=(6, 4))
sns.pointplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Renda média por posse de veículo')
plt.ylabel('Renda média')
plt.xlabel('Possui veículo')
plt.show()
````

-> Pessoas com veículo possuem uma renda maior do que quem não tem veículo.

-> As barras verticais são pequenas, o que indica que as médias estão mais concentradas e, portanto, sugerem estatísticas mais confiáveis

-> A diferença entre quem possui e não possui é ainda maior do que a do primeiro gráfico (posse_de_imovel), indicando que pode ser um indicativo ainda mais relevante da renda.
