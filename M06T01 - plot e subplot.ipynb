```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

for col in variaveis:
    sinasc[col].value_counts().sort_index().plot(kind='bar', color='skyblue')
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.grid(axis='y')
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sinasc.groupby('APGAR5')['CONSULTAS'].mean().plot(kind='bar', color='mediumseagreen')
plt.title('Média de Consultas por Valor do APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Média da Consultas')
plt.grid(axis='y')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

sinasc['SEXO'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 0], color='lightcoral', title='Sexo')
sinasc['GESTACAO'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 1], color='gold', title='Gestação')
sinasc['ESTCIVMAE'].value_counts().sort_index().plot(kind='bar', ax=axs[1, 0], color='steelblue', title='Estado Civil da Mãe')
sinasc['CONSULTAS'].value_counts().sort_index().plot(kind='bar', ax=axs[1, 1], color='limegreen', title='Consultas')

plt.tight_layout()
plt.show()
```
